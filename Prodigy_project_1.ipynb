```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel("Population Data.xls")
```

```python
dataset.head()
```

```python
import matplotlib.pyplot as plt


dataset.plot(x="Country Name", y="Total population_data", color="red", legend=False)
plt.title('Distribution of a Population')
plt.xlabel('Country Name')
plt.ylabel('Total Population')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
